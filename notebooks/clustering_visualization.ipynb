```python
import wordcloud
import matplotlib.pyplot as plt
import pandas as pd
```

```python
reviews_clustered = pd.read_parquet('../data/reviews_clustered.parquet')
```

```python
# visualize the clusters based on the words that fall into each cluster using wordclouds
wordclouds = []
for i in range(0,4):
    wordclouds.append(wordcloud.WordCloud(background_color="white", max_words=50).generate(" ".join(reviews_clustered[reviews_clustered["cluster"]==i]["reviewTokens"])))
    plt.imshow(wordclouds[i], interpolation='bilinear')
    plt.axis("off")
    print(f"Word cloud for cluster {i}")
    plt.show()
```
